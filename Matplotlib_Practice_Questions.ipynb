Plot the daily temperature over a week using a line chart.

```python
days =np.array(['MONDAY','TUESDAY','WEDNESDAY','THURSDAY','FRIDAY','SATURDAY','SUNDAY'])
temp =np.linspace(5,7,7)
temp

plt.plot(days,temp,color='red',marker='4',markersize='19')
plt.xlabel('days')
plt.ylabel('temp')
plt.grid()
plt.show()
```
